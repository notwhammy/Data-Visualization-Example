```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("MCI_2014_to_2018.csv")
df
```

```python
import folium as flm
df_new = pd.DataFrame(df['Neighbourhood'].value_counts()).reset_index()
df_new = df_new.rename(columns={'index':'District', 'Neighbourhood':'Count'})
df_new
```

```python
to_geo = 'toronto.geojson'


# create a plain world map
to_map = flm.Map(location=[43.6532, -79.3832], zoom_start=11)
t_scale = np.linspace(df_new['Count'].min(),
                      df_new['Count'].max(),
                      7, dtype=int)
t_scale[-1] = t_scale[-1] + 1
t_scale = t_scale.tolist() 

to_map.choropleth(
    geo_data=to_geo,
    data=df_new,
    columns=['District', 'Count'],
    key_on='feature.properties.FULLHOOD',
    threshold_scale=t_scale, 
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Auto Theft Crimes in Toronto',
    reset=True
)

to_map
```
