```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image # converting images into arrays

from wordcloud import WordCloud, STOPWORDS
diary = open("diaryaf.txt").read()
stopwords = set(STOPWORDS)

stopwords.add("said")

diary_mask = np.array(Image.open("afsil2.png"))

diary_wc = WordCloud(background_color='white', max_words=1000, mask=diary_mask)
diary_wc.generate(diary)
        
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(diary_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig("AnneFrank_mask.jpg")
```
